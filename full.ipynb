
## **расчет супа

```python
def get_portion(product):
    for datum in P.Food:
        p = datum[0]
        w = datum[1]
        if p.lower() == product.lower():
            return w
    raise ValueError('Продукт не найден', product)
```

```python
# калорийность на грамм (по материалам сайта http://www.calorizator.ru/)
calories = {'картофель': 0.76,
            'капуста': 0.27,
            'морковь': 0.32,
            'помидоры': 0.2,
            'зелень': 0.36,
            'лук': 0.47,
            'гречка': 3.3
           }
```

```python
80 * calories['гречка'] # калорийность порции гречки
```

```python
total = 0
for p in ['картофель', 'капуста', 'морковь', 'помидоры', 'зелень', 'лук']:
    w = get_portion(p)
    e = calories[p] * w
    print("%12s %5.0f %5.1f" % (p, w, e))
    total += e
total


```
